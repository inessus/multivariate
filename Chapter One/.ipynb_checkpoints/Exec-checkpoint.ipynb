```python
import torch 

import pandas as pd
import numpy as np
import matplotlib as mpl
import seaborn as sns
import matplotlib.pyplot as plt
```

```python

# exercise 1.1 
x1 = [3, 4, 2, 6, 8, 2, 5]
x2 = [5, 5.5, 4, 7, 10, 5, 7.5]

a = np.stack((x1, x2))

print("numpy:")
print(f"x1 mean: {a.mean(1)[0]:.2f} \nx2 mean: {a.mean(1)[1]:.2f}")
print("s11 std: {0:.2f} \ns12 std: {1:.2f}".format(a.std(1)[0], a.std(1)[1]))
print("x1 var: {0:.2f} \nx2 var: {1:.2f}".format(a.var(1)[0], a.var(1)[1]))
print("cov : {}".format(np.cov(a)))


b = torch.Tensor([x1, x2])
print("torch:")
print("x1 mean: {0:.3f} \nx2 mean: {1:.3f}".format(b.mean(1)[0], b.mean(1)[1]))
print("s11 std: {0:.3f} \ns12 std: {1:.3f}".format(b.std(1)[0], b.std(1)[1]))
print("x1 var: {0:.3f} \nx2 var: {1:.3f}".format(b.var(1)[0], b.var(1)[1]))
print("cov : {}".format(np.cov(b)))
```
